# Data Collection

```python
import pandas as pd
import numpy as np
```

```python
# Collects our data using the NBA API
from nba_api.stats.endpoints import leaguedashplayerstats
import time
all_data = []

for season in range(1997, 2026): # Change the years depending on what seasons you want to collect data from
    season_str = f"{season-1}-{str(season)[-2:]}"
    print(f"Fetching season {season}")
    
    stats = leaguedashplayerstats.LeagueDashPlayerStats(
        season=season_str,
        season_type_all_star='Regular Season',
        per_mode_detailed='PerGame'
    )

    df_season = stats.get_data_frames()[0]
    df_season["SEASON"] = season_str
    all_data.append(df_season)

    time.sleep(1)

df = pd.concat(all_data, ignore_index=True)

print("Done.")
print(df.tail())

# Optional line of code below to turn the dataframe into a CSV file. Change the file name and destination as needed.

# df.to_csv("../NBA Stats CSV/nba_data.csv", index=False)
```

# Data Cleaning

```python
# Clean our data
df = df.drop(['NICKNAME', 'TEAM_ID', 'W', 'L', 'W_PCT', 'REB', 'BLKA', 'PF', 'PFD', 'PLUS_MINUS', 'NBA_FANTASY_PTS', 'GP_RANK', 'W_RANK', 'L_RANK', 'W_PCT_RANK', 'MIN_RANK', 'FGM_RANK', 'FGA_RANK', 'FG_PCT_RANK', 'FG3M_RANK', 'FG3A_RANK', 'FG3_PCT_RANK', 'FTM_RANK', 'FTA_RANK', 'FT_PCT_RANK', 'OREB_RANK', 'DREB_RANK', 'REB_RANK', 'AST_RANK', 'TOV_RANK', 'STL_RANK', 'BLK_RANK', 'BLKA_RANK', 'PF_RANK', 'PFD_RANK', 'PTS_RANK', 'PLUS_MINUS_RANK', 'NBA_FANTASY_PTS_RANK', 'DD2_RANK', 'TD3_RANK', 'WNBA_FANTASY_PTS_RANK', 'TEAM_COUNT', 'DD2', 'TD3', 'WNBA_FANTASY_PTS'], axis=1)
df = df.dropna(how='any') # Removes stat columns we don't need and rows missing data
df = df[df['GP'] > 10] # Remove players with less than 11 games played in a season to prevent small sample sizes skewing data
```

# Feature Engineering

```python
# List of statistical categories used for fantasy projection
fantasyStats = ['FG_PCT', 'FG3M', 'FT_PCT', 'OREB', 'DREB', 
                    'AST', 'TOV', 'STL', 'BLK', 'PTS']

# Function to normalize a stat
def normalize(df, col):
    series = pd.to_numeric(df[col], errors='coerce')
    
    return (
        (series - series.min()) / (series.max() - series.min())
    )

# Function to automatically normalize all stats
def vorp(df):
    for col_name in fantasyStats:
        df = df.copy()
        df.loc[:, f"{col_name}_norm"] = normalize(df, col_name)
    return df
```

```python
# Normalize all the statistical categories in our dataframe
df = df.groupby(['SEASON']).apply(vorp)
```

# Model

```python
def player_comparison_tool(df, season, player):
    
    # Stops projection model if player does not exist
    if not ((df['SEASON'] == season) & (df['PLAYER_NAME'] == player)).any():
        print(f"Can't find player {player} in {season} season")
        return

    # Define the projected season
    start_year = int(season[:4]) # Grabs the first 4 characters (e.g., '2024' from '2024-25')
    next_start_year = start_year + 1
    next_end_year_str = str(next_start_year + 1)[-2:] # Grabs the last 2 digits of the following year
    proj_season = f"{next_start_year}-{next_end_year_str}"
    
    df = df.reset_index(drop=True)
    df = df.sort_values(['PLAYER_NAME','SEASON']) # Sorts the dataframe by player name's and seasons
    
    # Keep track of raw current stats for accurate delta calculations
    for col in fantasyStats:
        df["next_" + col] = df.groupby("PLAYER_NAME")[col].shift(-1)
        
    df_model = df.dropna(subset=["next_" + fantasyStats[0]])
    # Allows the player's past seasons to be used as comps, but excludes the current target season to prevent data leakage
    df_model = df_model[~((df_model["PLAYER_NAME"] == player) & (df_model["SEASON"] == season))]

    # Dynamic weighing system calculates which stats are most strongly predictive of future performance
    correlations = []
    for stat in fantasyStats:
        corr = df_model[stat].corr(df_model[f"next_{stat}"])
        correlations.append(abs(corr))
    
    feature_weights = np.array(correlations)
    feature_weights = feature_weights / feature_weights.sum()
    
    # X_norm is for distance (finding similar shapes)
    X_norm = df_model[[f"{s}_norm" for s in fantasyStats]].values
    # X_raw and Y_raw are for calculating actual basketball growth
    X_raw = df_model[fantasyStats].values 
    Y_raw = df_model[['next_' + s for s in fantasyStats]].values
    
    # Grab target's raw and normalized stats, plus age
    target_data = df[(df["PLAYER_NAME"] == player) & (df["SEASON"] == season)].iloc[0]
    target_norm = target_data[[f"{s}_norm" for s in fantasyStats]].values.astype(float)
    target_raw = target_data[fantasyStats].values.astype(float)
    player_age = target_data["AGE"]

    print(f"Projecting {player} for the {proj_season} season") # States which player is currently being projected
    
    player_weights = np.abs(target_norm)
    final_weights = feature_weights * player_weights
    final_weights = final_weights / final_weights.sum()

    # Calculate base statistical distance
    weighted_diff = (X_norm - target_norm) * final_weights
    base_distances = np.linalg.norm(weighted_diff, axis=1)
    
    # Quadratic Age Distance Penalty
    neighbor_ages = df_model["AGE"].values
    age_diff = np.abs(neighbor_ages - player_age)
    
    # Squaring the difference creates a curve:
    # 0 years diff = 0.000 penalty
    # 1 year diff  = 0.003 penalty (Barely noticeable)
    # 2 years diff = 0.012 penalty (Slight nudge)
    # 3 years diff = 0.027 penalty (Starting to push away)
    # 6 years diff = 0.108 penalty (Massive penalty, highly unlikely to match)
    
    age_multiplier = 0.003 # Tweak this single number to increase/decrease age importance
    age_penalty = (age_diff ** 2) * age_multiplier 
    
    # Calculate final distances combining stats and age
    final_distances = base_distances + age_penalty
    
    k = 25 # Total number of players used to project stats for current player, tweak this number to change number of players being used
    idx = np.argsort(final_distances)[:k]
    

    # Calculate weights for the weighted average of the neighbors
    sim_weights = 1 / (final_distances[idx] + 1e-6)
    sim_weights = sim_weights / sim_weights.sum() # Normalize weights
    
    # Delta math: Raw Next Season - Raw Current Season
    deltas = Y_raw[idx] - X_raw[idx]
    avg_delta = np.average(deltas, axis=0, weights=sim_weights)
    
    ## Apply the delta to the target's RAW stats, flooring all values at 0
    projection = np.maximum(0, target_raw + avg_delta)
    
    projection_dict = {
        "PLAYER_NAME": player,
        "PROJ_SEASON": proj_season
    }

    projection_dict.update(dict(zip(fantasyStats, np.round(projection, 2))))
    
    return projection_dict
```

# Evaluating the Model

```python
df_new = df[df['SEASON'].isin(['2023-24'])] # Sets to 2023-24 season to test model projections for the 2024-25 season
```

```python
# Creates an empty list and loops through the 2023-24 dataframe
final_projections = []
for baller in df_new.itertuples():
    projections = player_comparison_tool(df, baller.SEASON, baller.PLAYER_NAME) # Uses historical data and the function to create projections
    if (projections == None):
        continue
    final_projections.append(projections) # Projects every player in the 2023-24 dataframe for the next season and appends it to the list
test_projections = pd.DataFrame(final_projections)
```

```python
# Creates the 'df_real' dataframe containing the real player stats from the 2024-25 season
df_real = df[df['SEASON'].isin(['2024-25'])]
# Ensures both dataframes only contain players who played in both the 2023-24 and 2024-25 seasons
df_real = df_real[df_real['PLAYER_NAME'].isin(test_projections['PLAYER_NAME'])]
test_projections = test_projections[test_projections['PLAYER_NAME'].isin(df_real['PLAYER_NAME'])]

# # Optional line of code below to turn the dataframes into CSV files. Change the file name and destination as needed
# test_projections.to_csv('../NBA Stats CSV/projected_data.csv', index=False)
# df_real.to_csv('../NBA Stats CSV/real_data.csv', index=False)
```

```python
from sklearn.metrics import mean_squared_error

# Merge projections with real stats
merged_df = pd.merge(test_projections, df_real, on=['PLAYER_NAME'], suffixes=('_proj', '_real'))

mse_list = []
for stat in fantasyStats:
    real_vals = merged_df[f"{stat}_real"]
    
    # THE FIX: Added '_proj' so it looks for the correctly suffixed column
    proj_vals = merged_df[f"{stat}_proj"] 
    
    mse = mean_squared_error(real_vals, proj_vals)
    mse_list.append(mse)

overall_rmse = np.sqrt(np.mean(mse_list))
print(f"Overall Model RMSE: {overall_rmse:.3f}")
```

# Example Predictions

```python
# Create a visual comparison for a specific player
player_to_show = "LeBron James"

# Filter the merged dataframe for the specific player
example = merged_df[merged_df['PLAYER_NAME'] == player_to_show]

if example.empty:
    print(f"Could not find {player_to_show} in the testing dataframe.")
else:
    # Create a clean dataframe for display
    display_df = pd.DataFrame({
        'Stat': fantasyStats,
        # THE FIX: Added '_proj' to match the new column names
        'Projected': example[[f"{s}_proj" for s in fantasyStats]].values[0],
        'Actual (Real)': example[[f"{s}_real" for s in fantasyStats]].values[0]
    })

    print(f"--- 2024-25 Season Comparison: {player_to_show} ---")
    print(display_df.round(2).to_string(index=False))
```

# Converting to Fantasy Points

```python
# Define the fantasy point value for each statistical category
# Used Yahoo's H2H Points scoring system
def fantasy_pts_converter(row):
    pts = row['PTS']
    reb = row['OREB'] + row['DREB']
    ast = row['AST']
    blk = row['BLK']
    stl = row['STL']
    tov = row['TOV']
    return (pts - tov + 1.2*reb + 1.5*ast + 3*blk + 3*stl)
```

```python
merged_df['FPTS_proj'] = test_projections.apply(fantasy_pts_converter, axis = 1).values
merged_df['FPTS_real'] = df_real.apply(fantasy_pts_converter, axis = 1).values

fp_rmse = np.sqrt(mean_squared_error(merged_df['FPTS_real'], merged_df['FPTS_proj']))

print("--------------------------------------------------")
print(f"FANTASY POINTS RMSE: {fp_rmse:.2f}")
print("--------------------------------------------------\n")
```

```python
# We sort by the ACTUAL real fantasy points to see how the model projected the league's top performers
top_15_real = merged_df.sort_values(by='FPTS_real', ascending=False).head(15)

# Build a clean display dataframe
display_top15 = pd.DataFrame({
    'Player': top_15_real['PLAYER_NAME'],
    'Projected FPTS': top_15_real['FPTS_proj'].round(1),
    'Actual FPTS': top_15_real['FPTS_real'].round(1),
    'Diff (Proj - Act)': (top_15_real['FPTS_proj'] - top_15_real['FPTS_real']).round(1)
})

print("--- Top 15 Players by Actual Fantasy Points ---")
print(display_top15.to_string(index=False))
```
